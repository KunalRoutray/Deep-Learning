```python
# Imprt the libraries
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Input
from tensorflow.keras.callbacks import EarlyStopping
```

```python
#Input data (X) and target output(y)
X=np.array([[0,0],[0,1],[1,0],[1,1]])
y=np.array([[0],[1],[1],[0]])
```

```python
#Define the NN model

model = Sequential()
model.add(Input(shape=(2,)))
model.add(Dense(2,activation='sigmoid'))
model.add(Dense(1,activation='sigmoid'))

#Compile the Model
model.compile(loss='binary_crossentropy',metrics=['accuracy'])

#Train the model
early_stop=EarlyStopping(monitor='loss',patience=50)
model.fit(X,y,epochs=200,verbose=1,callbacks=[early_stop])


#need to train the model over 10000 epochs or declare a loop to train between 1,1000 and terminate when acc=1
```

```python
print(np.round(model.predict(X)))
```

```python
#Evaluate the model
loss,accuracy=model.evaluate(X,y)

print(f"Loss:{loss},Accuracy:{accuracy}")
```
