```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt
```

1.OBJECT DETECTION

```python
# Load the object detection model
model_detect = YOLO('yolov8n.pt')  # Use yolov8s.pt for better accuracy

# Run inference on an image
image_path ="https://cdn.pixabay.com/photo/2022/06/19/18/31/traffic-7272520_960_720.jpg"

# Replace with actual image path
results = model_detect(image_path)
results[0].save(filename='output_detection.jpg')

# Display result
img = cv2.imread('output_detection.jpg')
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("Object Detection")
plt.axis('off')
plt.show()
```

2.INSTANCE SEGMENTATION



```python
# Load the segmentation model
model_seg = YOLO('yolov8n-seg.pt')  # Segmentation model

# Run segmentation
results_seg = model_seg(image_path)
results_seg[0].save(filename='output_segmentation.jpg')

# Display result
img_seg = cv2.imread('output_segmentation.jpg')
plt.imshow(cv2.cvtColor(img_seg, cv2.COLOR_BGR2RGB))
plt.title("Instance Segmentation")
plt.axis('off')
plt.show()
```
